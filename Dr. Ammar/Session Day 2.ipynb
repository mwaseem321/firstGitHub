```python
#Steps of data visualization

#Step 1
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt

#step 2
#set a theme
sns.set_theme(style="ticks", color_codes=True)


#step 3. We can also import our own data
# import dataset
kashti= sns.load_dataset("titanic")
print(kashti)


# #step 4
# # plot basic graph with 1 variable
p= sns.countplot(x= "sex", data= kashti)
plt.show()
```

```python
# #step 5
# # plot basic graph with 2 variables
p= sns.countplot(x= "sex", data= kashti, hue="class")
plt.show()
```

```python
# #step 6
# # plot basic graph with 2 variables count plots with titles
p= sns.countplot(x= "sex", data= kashti, hue="class")

p.set_title("Title for kashti dataset")
plt.show()
```
