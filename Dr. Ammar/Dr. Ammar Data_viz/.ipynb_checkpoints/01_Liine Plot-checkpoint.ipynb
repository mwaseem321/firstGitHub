# import libraries

1-**seaborn automatically imports all the following libraries with itself*

- *numpy*

- *scify*

- *matplotlib*

- *pandas*

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

# Adding titles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")

plt.show()
```

## Adding x and y limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")
plt.xlim(2) #above plot x-axis started from 4.5, now it will start from 2 and similarly for the y-axis
plt.ylim(1)

plt.show()
```

## Set styles

We have the following styles

- darkgrid
- whitegrid
- dark
- white
- ticks

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")
sns.set_style("dark")
sns.set_style(style=None, rc=None)


plt.show()
```

## Size of figure

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool


#changing figure size
plt.figure(figsize=(10,8))

#draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")



plt.show()
```
